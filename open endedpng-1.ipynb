```python
# Convert timestamp and extract key temporal element# Convert timestamp and extract key temporal elements
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ns')
df['Hour'] = df['Timestamp'].dt.hour
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
df['IsWeekend'] = df['DayOfWeek'].isin([5, 6]).astype(int)

# Cyclical encoding for hour (better than raw hour for models)
df['Hour_sin'] = np.sin(2 * np.pi * df['Hour']/24)
df['Hour_cos'] = np.cos(2 * np.pi * df['Hour']/24)
```

```python
def create_sequences(data, time_steps=24):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps])
        y.append(data[i + time_steps, df.columns.get_loc("EnergyConsumption")])
    return np.array(X), np.array(y)

time_steps = 24  # 24 hours input to predict next hour
X, y = create_sequences(scaled_data, time_steps)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

model = Sequential()

# First LSTM layer with return_sequences=True (because another LSTM follows)
model.add(LSTM(64, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2))

# Second LSTM layer
model.add(LSTM(32, return_sequences=False))
model.add(Dropout(0.2))

# Dense layers for richer learning
model.add(Dense(16, activation='relu'))
model.add(Dense(1))  # Output layer (regression)

# Compile model
model.compile(optimizer='adam', loss='mse')
model.summary()

```

```python
history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X)

# Plot actual vs predicted
plt.figure(figsize=(14, 6))
plt.plot(y[:200], label="Actual")
plt.plot(y_pred[:200], label="Predicted")
plt.xlabel("Time Steps")
plt.ylabel("Scaled Energy Consumption")
plt.title("LSTM Energy Forecasting")
plt.legend()
plt.grid(True)
plt.show()

# Evaluate MSE
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```
